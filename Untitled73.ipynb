CSV Файлын Оқу және Тізімге Сақтау

```python
from os import read
with open("/content/Salaries.csv", "r", encoding="utf-8") as file:
  read = file.readlines()
  data_rows = read[1:11]
print(len(data_rows))
print(data_rows[0])
```

Кортеж және Сөздік – Алғашқы 3 Адамның Атын Сақтау


```python
first_three_names = []
for i in range(3):
  #if i < len(data_rows):
    name = data_rows[i].split(",")[1]
    first_three_names.append(name)
print(first_three_names)
```

```python
job_titles = []
for i in range(3):
    job= data_rows[i].split(",")[2]
    job_titles.append(job)
print(job_titles)
```

2 декертерді бөлек алып алып оларды дикшнариға айналдырдым, ол үшін dict(zip қолдандым

```python

employee_titles = dict(zip(first_three_names, job_titles))
print(employee_titles)

```

Тізімдік Өрнектер (List Comprehension)

```python
#data_rows
BasePay = []
for n in data_rows:
  columns =n.split(",")
  BasePay.append(columns[3])
print(BasePay)
```

```python
base_pays_str = [n.split(",")[3] for n in data_rows]
print(base_pays_str)
```

Жиын (Set) – Бірегей Атақтарды Табу

```python
jobTitle = []
for n in data_rows:
  columns =n.split(",")
  jobTitle.append(columns[2])
print(jobTitle)
```

```python
unique = set(jobTitle)
print(unique)
```
